```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_excel("Employee data.xlsx")
```

```python
rel_freq=data['gender'].value_counts()/len(data)
```

```python
rel_freq
```

```python
precentage = rel_freq*100
```

```python
precentage
```

```python
freq = data['gender'].value_counts()
```

```python
# combine all in one table
gender_table = pd.DataFrame({
    "Gender": freq.index,
    "Frequency": freq.values,
    "Relative Frequency": rel_freq.round(3).values,
    "Precentage %": precentage.round(2).values
})
```

```python
gender_table
```

```python
# Add Total row
total_row = pd.DataFrame({
    "Gender": ["Total"],
    "Frequency": [freq.sum()],
    "Relative Frequency": [rel_freq.sum().round(3)],
    "Precentage %": precentage.sum().round(2)
})

gender_table = pd.concat([gender_table, total_row], ignore_index=True)

#display table
print(gender_table)
```

```python
# pie chart
plt.figure(figsize=(6, 6))
plt.pie(freq, labels=freq.index, autopct='%1.1f%%', startangle=90)
plt.title('Gender')
plt.show()
```

```python
# bar chart
plt.figure(figsize=(6, 6))
plt.bar(freq.index, freq.values, color= ['blue','red'])
plt.title('Gender')
plt.xlabel('Gender')
plt.ylabel('Frequency')
plt.show()
```

```python
# number of classes
number_classes = 5
```

```python
# claculate class width
salary_min = data['salary'].min()
salary_max = data['salary'].max()
class_width = (salary_max - salary_min) / number_classes
```

```python
class_width
```

```python
# create bins and lables
bins = [salary_min + i * class_width for i in range(number_classes + 1)]
```

```python
bins
```

```python
labels = [f"{int(bins[i])} - {int(bins[i+1])}" for i in range(number_classes)]
```

```python
labels
```

```python
# categorize salary into classes
data['salary_class'] = pd.cut(data['salary'], bins=bins, labels=labels, include_lowest=True)
```

```python
data['salary_class']
```

```python
#freq, relative frequence, percentage
freq = data['salary_class'].value_counts().sort_index()
rel_freq = freq / len(data)
precentage = rel_freq * 100
```

```python
freq
```

```python
# put in a table
freq_table = pd.DataFrame({
    "Salary Class": freq.index,
    "Frequency": freq.values,
    "Relative Frequency": rel_freq.round(3).values,
    "Percentage %": precentage.round(2).values
})
```

```python
freq_table
```

```python
# add total row
total_row = pd.DataFrame({
    "Salary Class": ["Total"],
    "Frequency": [freq.sum()],
    "Relative Frequency": [rel_freq.sum().round(3)],
    "Percentage %": precentage.sum().round(2)
})

freq_table = pd.concat([freq_table, total_row], ignore_index=True)

print(freq_table)
```

```python
# plot histogram
plt.figure(figsize=(6, 6))
plt.hist(data["salary"], bins=bins, color="skyblue", edgecolor="black")
plt.xlabel("Salary Classes")
plt.ylabel("Frequency")
plt.title("Histogram of Salary")
plt.xticks(bins, rotation=45)
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()
```




# Create Boxplot

```python
# create Boxplot
plt.figure(figsize=(6, 6))
plt.boxplot(
    data["salary"],
    vert=True,
    patch_artist=True,
    boxprops=dict(facecolor="lightblue", color="black"),
    medianprops=dict(color="red", linewidth=2),
    whiskerprops=dict(color="black"),
    capprops=dict(color="black"),
    flierprops=dict(marker="o", color="orange", alpha=0.6)
)

# add title and labels
plt.title("Boxplot of Salary", fontsize=12)
plt.ylabel("Salary", fontsize=11)
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()
```
